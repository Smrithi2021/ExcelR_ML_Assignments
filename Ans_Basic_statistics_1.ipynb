# To visualize the distribution and relationship of numerical and categorical variables in the dataset.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
from scipy.stats import skew
```

```python
read_df.head()
```

```python
numerical_cols=read_df.select_dtypes(include=['float64','int64']).columns
```

```python
numerical_cols
```

```python
for col in numerical_cols:
    plt.figure(figsize=(8,5))
    col_skewness = skew(read_df[col].dropna())
    
    sns.histplot(read_df[col], bins=30, kde=True)
    plt.title(f'Histogram of {col}(Skewness = {col_skewness:.2f})')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    
    plt.show()
```

```python
#Inference-

# All the histograms except the one on Discount rate % are right skewed. It means the data points are concentrated towards
# the left with a few larger values stretching out to the right called as outliers, in most cases (Volume, Avg Price, 
# Total Sales Value, Discount Amount and Net Sales Value). In case of Discount rate% the outliers mostly seem to be 
# concentrated in left however since the difference in between max and min values are less they cannot be called as outliers in general
# but rather the data is almost equivalently distributed.
```

```python
for col in numerical_cols:
    plt.figure(figsize=(8,5))
    sns.boxplot(x=read_df[col])
    Q1= read_df[col].quantile(0.25)
    Q3= read_df[col].quantile(0.75)
    IQR = Q3-Q1
    
    plt.title(f'Boxplot of {col} \n IQR={IQR:.2f}')
    plt.xlabel(col)
    plt.show()
```

```python
# # Inference-

# The outliers are mostly concentrated towards the right side in all cases except Discount Rate %. 65% of data lies in 
# between the IQR. For Discount Rate % although the outliers lie in the left side and majority of the data points lie
# in between 14 and 18 the number of outliers are less compared to other paramters as this expands upto 2 units beyond the 
# whiskers.
```

# Conversion of Categorical Data into Dummy Variables

```python
# One-hot encoding is a common technique used to convert categorical data into a numerical format 
# that can be easily understood by machine learning algorithms. Machine learning algorithms often 
# rely on mathematical operations that are not defined for categorical data. 

# One-hot encoding transforms these categorical values into numerical representations that can be used in calculations.

# If categorical data is represented as integers, the algorithm might assume an ordinal relationship between the categories. 
# For instance, if "red" is encoded as 1 and "blue" as 2, the algorithm might incorrectly assume that "blue" is greater than 
# "red." One-hot encoding prevents this by creating separate binary features for each category.

# One-hot encoding can help improve the performance of machine learning models, especially when dealing with
# categorical features that have a significant impact on the target variable.  By representing these 
# features as numerical values, the model can learn their relationships with the target more effectively.

# In one-hot encoding, each category is represented as a binary vector. The vector has a length equal to the number of 
# unique categories, and only one element in the vector is 1, while the others are 0. This indicates the presence or absence
# of that particular category.

# Consider a categorical feature "color" with three possible values: "red", "blue", and "black". After encoding the data will
# be represented as below.
    
# Red- [1,0,0]
# Blue- [0,1,0]
# Black- [1,0,0]

```

```python
categorical_columns=new_df.select_dtypes(include=['object','category']).columns
categorical_columns
```
